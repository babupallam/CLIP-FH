## Abstract

üìå **Main Idea:**  
This paper proposes a new **hand-based person recognition** method, called **MBA-Net**, for criminal investigations where only hand images are available (e.g., sexual abuse cases).

üìå **Key Features of MBA-Net:**  
‚úÖ **Multi-Branch Architecture** ‚Äì Uses **multiple branches** to extract different features from the hand image.  
‚úÖ **Attention Mechanisms** ‚Äì Includes **channel and spatial attention modules** to focus on important hand details and ignore background noise.  
‚úÖ **Global Feature Learning** ‚Äì Also includes a **global branch** (without attention) to capture overall hand structure.  
‚úÖ **Improved Spatial Understanding** ‚Äì Uses **relative positional encodings** in the spatial attention module to understand pixel locations better.  

üìå **Performance:**  
- Tested on **two large multi-ethnic public hand datasets**.  
- **Achieves state-of-the-art results**, outperforming existing hand-based identification methods.  
- Source code is available for replication: **[GitHub Link](https://github.com/nathanlem1/MBA-Net)**.  

üìå **Key Contribution:**  
This method **improves hand-based person recognition** using deep learning, making it **more accurate and reliable for criminal investigations** where hands are the only evidence.

## **Introduction**
üìå **Main Idea:**  
Biometric identification using **hand images** is gaining attention for **person recognition**, especially in **criminal investigations** where hands might be the only available evidence (e.g., sexual abuse cases).  

---

### **1Ô∏è‚É£ Why Hand Images for Person Recognition?**  
‚úÖ Hand images have **strong and stable features** after adulthood.  
‚úÖ Compared to other biometric traits (face, iris, fingerprints), **hand images are more stable** and have **less variability** over time.  
‚úÖ Digital cameras can **capture hand images easily**, making them useful for **real-world criminal investigations**.  

---

### **2Ô∏è‚É£ Previous Approaches & Their Limitations**  
üîπ **CNN + SVM Approach:**  
- Earlier works **trained CNNs on hand images** and extracted features for classification using **Support Vector Machines (SVMs)**.  
- Some methods used **additional modalities like Near-Infrared (NIR) images** to improve performance.  
‚ùå **Limitation:** These methods are **not end-to-end**, meaning they require **separate feature extraction and classification steps**.  

üîπ **Self-Attention in Deep Learning:**  
- **Transformers** use **self-attention mechanisms** to capture long-term interactions across pixels, channels, and feature maps.  
- Attention-based person Re-ID methods ([7], [8]) have been successful, but they **focus on full-body images**.  
‚ùå **Limitation:** Applying attention **across all network layers is computationally expensive**, especially for large image dimensions.  

üîπ **Problem with Self-Attention:**  
- Attention mechanisms **do not naturally model position information**.  
- Relative positional encoding was introduced in **NLP ([9])** and later adapted for vision tasks ([10]).  
‚ùå **Limitation:** Existing methods use attention mechanisms **without considering pixel positions**, leading to **equivariance to pixel shuffling** (i.e., position confusion).  

---

### **3Ô∏è‚É£ The Proposed MBA-Net Model**  
üöÄ **Solution: Multi-Branch with Attention Network (MBA-Net)**  
A novel **hand-based person recognition model** that:  
‚úÖ Uses **channel & spatial attention** in separate branches **(multi-branch design)**.  
‚úÖ Includes **a global branch** (without attention) to **capture full hand structure**.  
‚úÖ Incorporates **relative positional encodings** to improve **spatial feature learning**.  
‚úÖ Applies attention modules **only to low-resolution feature maps** for **better computational efficiency**.  
‚úÖ Works **end-to-end** without requiring **separate feature extraction & classification steps**.  

---

### **4Ô∏è‚É£ Contributions of This Work**  
üìå **Key Innovations:**  
1Ô∏è‚É£ **Multi-Branch Attention Network**: Combines **channel & spatial attention** with a global branch for **efficient hand-based recognition**.  
2Ô∏è‚É£ **Relative Positional Encoding**: Captures **spatial positions of pixels** to address **attention mechanism limitations**.  
3Ô∏è‚É£ **State-of-the-Art Evaluations**: Tested on **two large multi-ethnic hand datasets** (**11k Hands** & **HD dataset**).

## **3Ô∏è‚É£ Network Architecture Overview**
üéØ **Key Idea:**  
- **Combines attention modules with ResNet50 backbone** in a **multi-branch design** for efficient hand-based recognition.  

üõ† **Architecture Details:**  
- **Uses ResNet50** (pre-trained on ImageNet) as the base model.  
- **Three parallel branches**:  
  1Ô∏è‚É£ **Channel Attention Branch (CAM)**  
  2Ô∏è‚É£ **Spatial Attention Branch (SAM-RPE)**  
  3Ô∏è‚É£ **Global Feature Branch (no attention, captures overall hand structure)**  
- **Each branch has**:  
  ‚úÖ Fully Connected (FC) layer  
  ‚úÖ Batch Normalization (BN)  
  ‚úÖ Leaky ReLU (LReLU)  
  ‚úÖ Dropout (0.5) to reduce overfitting  
- **Final Feature Representation**:  
  - During **training**, each branch makes an independent **ID prediction** (using Cross-Entropy Loss).  
  - During **testing**, the feature maps from all **three branches are concatenated** into a **6144-D feature vector** and compared using **cosine distance** for person recognition.  

‚úÖ **Key Benefit:**  
- **More efficient computation** (uses attention only in later layers).  
- **Robust recognition** (combines local and global features).  
- **End-to-end training** (unlike previous CNN+SVM methods).  

---

### **üìù Key Takeaways**
‚úÖ **MBA-Net improves accuracy by combining channel & spatial attention modules.**  
‚úÖ **Relative Positional Encoding helps maintain spatial consistency for hand images.**  
‚úÖ **Uses ResNet50 but can be adapted to other networks like Inception.**  
‚úÖ **Multi-branch design balances computational efficiency & feature richness.**

### **Explanation of the Figure: Structure of MBA-Net**

üìå **Main Idea:**  
This figure illustrates the **architecture of MBA-Net**, showing how **hand images are processed** through different branches to extract features for **hand-based person identification**.

---

### **üîπ Step-by-Step Breakdown of the Figure**

#### **1Ô∏è‚É£ Input: Hand Image**
- The **input image** (hand image) has **height (H) and width (W)**.
- This image is fed into the **Backbone Network (ResNet50)**.

---

#### **2Ô∏è‚É£ Backbone Network (Feature Extractor) ‚Äì ResNet50**
- **Purpose:** Extract important **low-level and high-level features** from the hand image.
- The ResNet50 **pretrained on ImageNet** is used up to **layer 3** (meaning the first three layers are unchanged).
- The extracted feature map is then **passed into three separate branches**.

---

#### **3Ô∏è‚É£ Multi-Branch Network**
After passing through ResNet50, the model **splits into three branches**:

‚úÖ **1st Branch: Spatial Attention (SAM-RPE) ‚Äì S3, L4, S4**  
   - **Focuses on spatial regions** in the hand image (important pixels).  
   - Includes **GAP (Global Average Pooling)**, followed by a **fully connected (FC) layer**, batch normalization (BN), leaky ReLU (LReLU), and dropout.  
   - Produces a **2048-dimensional feature vector** (**vector s** ‚Üí **vector s'**).  

‚úÖ **2nd Branch: Global Feature Branch ‚Äì L4**  
   - Extracts **general features** without attention mechanisms.  
   - Includes **GAP, FC, BN, LReLU, and Dropout**.  
   - Produces a **2048-dimensional feature vector** (**vector g** ‚Üí **vector g'**).  

‚úÖ **3rd Branch: Channel Attention (CAM) ‚Äì C3, L4, C4**  
   - **Identifies the most important feature channels** from the image.  
   - Similar to the other branches, it includes **GAP, FC, BN, LReLU, and Dropout**.  
   - Produces a **2048-dimensional feature vector** (**vector c** ‚Üí **vector c'**).  

---

#### **4Ô∏è‚É£ Final Identity Prediction**
- Each branch independently **predicts the identity of the hand image**.
- The **fully connected (FC) layers** in each branch output identity predictions.
- In the final model:
  - During training, all three branches contribute **independently**.
  - During testing, all feature vectors (**s, g, c**) are concatenated into a **single 6144-dimensional feature vector** for final identification.

---

### **üîπ Key Takeaways from the Figure**
‚úÖ The **hand image** is processed using **ResNet50 as the backbone**.  
‚úÖ Three **parallel branches** extract different types of features:  
   - **Spatial Attention (S3, S4)** ‚Üí Highlights important regions.  
   - **Global Features (L4)** ‚Üí Extracts overall hand structure.  
   - **Channel Attention (C3, C4)** ‚Üí Focuses on the most important feature channels.  
‚úÖ Each branch outputs **2048-dimensional feature vectors**, which are **used to predict identity**.  
‚úÖ **Final feature fusion** (concatenation) is done during testing for the best accuracy.  

---

### **üìå Why This Architecture?**
- **More Accurate:** Each branch extracts **complementary information** for better identification.  
- **More Robust:** Attention mechanisms **filter out noise and background** while focusing on hand features.  
- **Computationally Efficient:** Uses attention only at later layers, reducing processing cost.

## **Attention Modules (CAM and SAM-RPE) in MBA-Net**

image.png

### **Explanation of Fig. 1: Attention Modules (CAM and SAM-RPE) in MBA-Net**

üìå **Main Idea:**  
This figure explains the **two attention mechanisms** used in the MBA-Net model:
1. **Channel Attention Module (CAM)**
2. **Spatial Attention Module with Relative Positional Encoding (SAM-RPE)**  

Both modules help the model focus on **important features in hand images** while ignoring background noise.

---

## **üîπ (a) Channel Attention Module (CAM)**
üéØ **Purpose:**  
- Identifies **which feature channels** are the most important for recognizing the hand.  
- Helps the model **focus on relevant feature maps** and ignore unnecessary ones.  

üõ† **How it Works:**
1Ô∏è‚É£ The input feature map **\(E_i\)** (size: **C √ó H √ó W**) is processed into three matrices:
   - **K (Keys)**
   - **Q (Queries)**
   - **V (Values)**  
   (Each has the same shape: **C √ó HW**)  

2Ô∏è‚É£ The **dot-product of Q and K** is computed to find the **channel-wise relationships**, producing an **attention map \(A_c\) (C √ó C matrix).**  

3Ô∏è‚É£ The attention map is then **multiplied with V (Values)** to weight the most important feature channels.  

4Ô∏è‚É£ The updated feature map is **reshaped back to C √ó H √ó W** and **added to the original input (\(E_i\))** to produce the final output \(E_o\).  

‚úÖ **Key Benefit:**  
- Helps the network **identify the most meaningful feature channels** in hand images.  

---

## **üîπ (b) Spatial Attention Module with Relative Positional Encoding (SAM-RPE)**
üéØ **Purpose:**  
- Identifies **which regions (pixels) in the image are most important**.  
- Helps the model **understand spatial relationships** between different parts of the hand.  

üõ† **How it Works:**
1Ô∏è‚É£ The input feature map \(E_i\) (**C √ó H √ó W**) is processed into:
   - **K (Keys)**
   - **Q (Queries)**
   - **V (Values)**  
   (But here, Keys and Queries have smaller dimensions: **\(d_k \times HW\)**, where \(d_k = C/8\) to reduce computation.)  

2Ô∏è‚É£ The **dot-product of Q and K** is computed to find the **spatial attention map \(A_s\) (HW √ó HW matrix).**  

3Ô∏è‚É£ **Relative Positional Encoding (RPE) is applied** to fix a common problem in self-attention:
   - The network **adds learned position embeddings** for height (\(P_H\)) and width (\(P_W\)), which helps maintain **spatial order** in the image.  
   - Without this, attention might treat different hand parts as interchangeable, leading to poor recognition.  

4Ô∏è‚É£ The **final self-attended feature map** is obtained by combining:
   - The **weighted values V**
   - The **relative position-enhanced features from \(P_H\) and \(P_W\)**  

5Ô∏è‚É£ The **output is normalized (BN: Batch Normalization)** and added back to the **original feature map** to get \(E_o\).  

‚úÖ **Key Benefit:**  
- Helps the model **preserve the spatial structure of the hand** (e.g., keeping the correct positions of fingers and palm).  

---

### **üìù Key Takeaways from the Figure**
‚úî **Channel Attention (CAM)** ‚Üí Finds **which features are important**  
‚úî **Spatial Attention (SAM-RPE)** ‚Üí Finds **where the important features are**  
‚úî **Relative Positional Encoding (RPE)** ‚Üí Helps maintain **spatial consistency**  
‚úî **Final Output** ‚Üí More **accurate hand-based person identification**

## **Implementation of the Proposed System (MBA-Net)**

---

### **üìå Step 1: Understanding the Purpose of MBA-Net**
- **Goal:** Identify individuals using **hand images** for biometric recognition, particularly for **criminal investigations**.
- **Challenge:** Hand images have **background noise**, and important features (like finger structure, palm patterns) must be **highlighted** while ignoring irrelevant details.
- **Solution:** **MBA-Net**, a deep learning model that **extracts hand features using attention mechanisms** while improving efficiency.

---

### **üìå Step 2: How MBA-Net Works**
MBA-Net has three main components:
1Ô∏è‚É£ **Backbone Network (Feature Extractor) ‚Äì ResNet50**  
2Ô∏è‚É£ **Attention Modules ‚Äì Channel & Spatial Attention**  
3Ô∏è‚É£ **Final Classification for Identity Prediction**  

Each component is **trained end-to-end** to automatically learn important features from hand images.

---

### **üñ• Step 3: Implementation Details**
#### **üîπ 1. Backbone Network (Feature Extractor)**
- Uses a **pretrained ResNet50** network as a starting point.
- ResNet50 is a **deep convolutional neural network (CNN)** that processes hand images and extracts **basic feature maps**.
- Only the **first three layers of ResNet50** remain unchanged.
- After the third layer, the network **branches into three parts**.

---

#### **üîπ 2. Attention Mechanisms**
MBA-Net **improves feature extraction** using two types of **attention mechanisms**:

1Ô∏è‚É£ **Channel Attention Module (CAM) ‚Äì Identifies Important Feature Channels**
   - **Purpose:** Focus on **which features are important** for hand recognition.
   - **How it Works:**
     - The network breaks down the **feature maps into separate channels**.
     - It **assigns more importance** to channels that contain **relevant hand details**.
     - Uses **dot-product attention** to calculate the relationship between channels.
     - The model **learns to enhance important channels** while reducing the effect of less useful ones.

2Ô∏è‚É£ **Spatial Attention Module with Relative Positional Encoding (SAM-RPE) ‚Äì Identifies Important Hand Regions**
   - **Purpose:** Focus on **where in the image the most important information is**.
   - **Problem with Normal Attention:** Traditional spatial attention does not consider **pixel positions**.
   - **Solution:** **Relative Positional Encoding (RPE)**
     - Adds extra information to **keep track of where each pixel is**.
     - This prevents **loss of spatial relationships** when images are shifted or resized.

‚úÖ **Benefit of Using Attention:**  
- Helps the model **filter out background noise**.  
- Focuses on **hand structure** while ignoring accessories, shadows, or unnecessary details.  

---

#### **üîπ 3. Multi-Branch Design (Final Classification)**
After passing through **attention modules**, the model **splits into three branches**:
1Ô∏è‚É£ **Channel Attention Branch (CAM output)**  
2Ô∏è‚É£ **Spatial Attention Branch (SAM-RPE output)**  
3Ô∏è‚É£ **Global Feature Branch (Normal ResNet50 Features, No Attention)**  

üìå **Each branch processes the hand image separately** and produces **a 2048-dimensional feature vector**.

- These vectors are then **combined into a single 6144-dimensional feature vector**.  
- Finally, a **fully connected (FC) layer with a softmax function** predicts the **identity of the person**.

3Ô∏è‚É£ **Testing Phase:**
